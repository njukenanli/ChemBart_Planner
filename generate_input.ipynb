```python
import json
from rdkit import Chem
from rdkit.Chem import rdChemReactions, Draw
import matplotlib.pyplot as plt
import pickle
import re
from tqdm import tqdm 
import pickle
import pandas as pd


def checkretromol(pre_data, number_routes=1, chirality=True):
    mols = pre_data
    label = number_routes
    print('Original smiles list has {} data, start checking'.format(len(mols)))

    valid_mols = []  # 用于存储有效的分子
    #valid_labels = []  # 用于存储对应的有效标签

    for i in range(len(mols)):
        if mols[i] == "":
            continue
        mol = Chem.MolFromSmiles(mols[i])
        if mol is not None:
            if chirality == False:
                Chem.AssignStereochemistry(mol, cleanIt=True, force=True)
                Chem.DetectBondStereochemistry(mol, -1)
                Chem.AssignAtomChiralTagsFromStructure(mol, -1)
                Chem.RemoveStereochemistry(mol)
            # rdkit 检测分子正确性
            mol = Chem.MolToSmiles(mol, canonical=True, isomericSmiles=True, kekuleSmiles=False)
            valid_mols.append(mol)
            #valid_labels.append(int(labels[i]))

    # 提取表达式和标签
    data = [(mol, label) for mol in valid_mols]
    print('Finish checking molecules, now has {} data'.format(len(data)))
    return data

def mol_convert(pre_data, chirality=True):
    mol = pre_data


    mol = Chem.MolFromSmiles(mol)
    if mol is not None:
        if chirality == False:
            Chem.AssignStereochemistry(mol, cleanIt=True, force=True)
            Chem.DetectBondStereochemistry(mol, -1)
            Chem.AssignAtomChiralTagsFromStructure(mol, -1)
            Chem.RemoveStereochemistry(mol)
        # rdkit 检测分子正确性
        mol = Chem.MolToSmiles(mol, canonical=True, isomericSmiles=False, kekuleSmiles=False)

    return mol

```

```python
mol = 'COC(=O)c1ccc2c(c1)NC(=O)/C2=C(\\Nc1ccc(N(C)C(=O)C(C)N2CCN(C)CC2)cc1)c1ccccc1'
mol_convert(mol)
```

## JMC

```python
file_path = 'JMC/JMC_benchmark.csv'

file = pd.read_csv(file_path)



mollist =  file['SMILES'].tolist()
tasklist = checkretromol(mollist,number_routes=1,chirality=True)

#print(tasklist)
```

```python
file['SMILES_pro'] = file['SMILES'].apply(lambda x: mol_convert(x, chirality=True))
file.to_csv('JMC/JMC_benchmark.csv', index=False)
```
